```python
#memasukkan data, mengganti nama record, menggeser value column
import pandas as pd
from numpy import math

#memasukkan data
df = pd.read_excel ('D:\MyProject\RekomendasiBeasiswa\Data\BeasiswaBI.xlsx')
df
```

```python
#Menghapus baris pertama 
df=df.drop([0])

#mengganti nama record
df=df.rename(columns={
    'NR':'NR1',
    'Unnamed: 2':'NR2',
    'Unnamed: 3':'NR3',
    'Unnamed: 4':'NR4',
    'Unnamed: 5':'NR5',
    'Unnamed: 6':'NR6',
    'Unnamed: 7':'NR7',
    'Unnamed: 8':'NR8',
    'Unnamed: 9':'NR9',
    'Unnamed: 10':'NR10',
    'Unnamed: 11':'NR11'
},inplace=False)
header=list(df)

#Menggeser value column
for i in df.index:
    j=1
    while math.isnan(float(df.loc[i, header[j]])):
        temp=df.loc[i, header[j]]
        for k in range(0,11):
            a=j+k
            b=a+1
            df.loc[i, header[a]]=df.loc[i, header[b]]
        df.loc[i, header[11]]=temp
df=df.reset_index(drop=True)
df
```

```python
#menghapus column
# df.info()
for i in range(8,12):
    if df[header[i]].isnull().sum():
        del df[header[i]]
df.info()
```

```python
#Mengisi data yang kosong
x=df.loc[:,header[0]]
del df['Status Beasiswa BI']
df = df.apply(pd.to_numeric)
df['Decision']=x
# del df['NR1']
# del df['NR2']
# del df['NR3']
# del df['NR4']
# del df['NR5']
del df['NR6']
del df['NR7']
df=df.reset_index(drop=True)
df
```

## Splitting Data

```python
import numpy as np
from sklearn.model_selection import train_test_split
##80 train, 20 test
df_train, df_test=np.split(df.sample(frac=1,random_state=42),[int(0.8*len(df))])

##80 train,10 validate,10 test
# df_train, df_validate, df_test = np.split(df.sample(frac=1,random_state=42),[int(.8*len(df)),int(.9*len(df))])
```

```python
from chefboost import Chefboost as chef
import pandas as pd

config = {'algorithm': 'C4.5'}
model = chef.fit(df_train, config = config)
```

```python
## CEK KETEPATAN DALAM KLASIFIKASI
decision_predict=[]
TP=0
TN=0
FP=0
FN=0

for index,instance in df_test.iterrows():
#     print(instance)
    prediction = chef.predict(model, instance)
#     print(index,instance['Decision'],prediction)
#     decision_predict.append(prediction)
    if instance['Decision'] != prediction:
        if (instance['Decision'] == "Iya") and (prediction !="Iya"):
            FN+=1
        else:
            FP+=1
    else:
        if (instance['Decision'] == "Iya") and (prediction =="Iya"):
            TP+=1
        else:
            TN+=1
    print(index," ",instance['Decision'],prediction,"| TP=",TP,", TN=",TN,", FP=",FP, ", FN=",FN)
        
# df_test['Decision Predict']=decision_predict
```
